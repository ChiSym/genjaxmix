```python
from genjaxmix.rejuvenation import gibbs_move, propose_parameters
from genjax._src.core.interpreters.incremental import Diff

key = jax.random.key(3)
model = generate(100)
model_args = (1.0, 0.0, 1.0, 3.0, 0.5)
tr = model.simulate(key, model_args)
obs = tr.get_choices()("assignments")

# print("Data ", obs[:, "y2"])
# print("c = ", obs[:,"c"])
print("Old parameters")
print("mu ", tr.get_choices()["hyperparameters", "mu"])
print("sigma ", tr.get_choices()["hyperparameters", "sigma"])
print("logp ", tr.get_choices()["hyperparameters", "logp"])
key, *subkeys = jax.random.split(key, 5)
tr_new = gibbs_move(model, propose_parameters, model_args, tr, obs, subkeys[0])
tr_new = gibbs_move(model, propose_parameters, model_args, tr_new, obs, subkeys[1])
tr_new = gibbs_move(model, propose_parameters, model_args, tr_new, obs, subkeys[2])
tr_new = gibbs_move(model, propose_parameters, model_args, tr_new, obs, subkeys[3])
# print(tr_new)
print()
print("New parameters")
print("mu ", tr_new.get_choices()["hyperparameters", "mu"])
print("sigma ", tr_new.get_choices()["hyperparameters", "sigma"])
print("logp ", tr_new.get_choices()["hyperparameters", "logp"])
```
